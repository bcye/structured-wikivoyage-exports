### Doc structure

* curr: the element being worked on
* parent: one level up
* root: reference to the root

each has a type, properties and children (tree)

### Parsing Text
Will start or continue a text block, a text block will be markdown and support (and parsed into) from stylistic tags, internal (wiki) and external links and text

### Stylistic Tags

i.e. Bold, Italics, etc. We expressly don't support very weird edge cases like `test **test2 *test3 test4** test5*` for now (we could run an analysis if they even exist for giggles i guess)

```python
init_root = {
    'type': "root",
    'properties': {},
    'children': [],
}
init_curr = root
init_parent = None

def parse(doc: mwp.wikicode.Wikicode, root=init_root, curr=init_curr, parent=init_parent):
    # if curr: goes level deeper, else could for example make a sibling
    def new_el(for_parent, type):
        parent = for_parent
        curr = {
            'type': type,
            'properties': {},
            'children': []
        }
        parent['children'].append(curr)
    
    def handle_txt(node: nodes.text.Text):
        val: str = node.value
        if curr['type'] == 'text':
            # already working on a text, append
            curr['properties']['value'] = curr['properties']['value'] + val
        else:
            # not yet at a text
            new_el('text')
            curr['properties']['value'] = val

    def handle_tag(node: nodes.tag.Tag):
        # assert that we are in a text element (should the text start with a tag)
        if curr['type'] != "text":
            handle_txt(nodes.text.Text(""))
        
        wrap = ""
        match node.tag:
            case 'b':
                wrap = "**"
            case _:
                print(node.tag)
                exit()
        
        # start the wrapper tag, recursively parse the children
        curr['properties']['value'] += wrap
        parse(node.contents, root, curr, parent)
        if curr['type'] != "text":
            print("panic, not in text after recursive parsing")
            exit()
        curr['properties']['value'] += wrap
    
    def handle_heading(node: nodes.heading.Heading):
        # find level of last heading of equal greatness, make sibling
        new_el()
```
